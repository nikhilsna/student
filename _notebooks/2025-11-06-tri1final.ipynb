# Trimester 1 Final Presentation

## (1) MCQ Review & Performance Analysis

I scored 34/42 on the Practice Exam, but this was mostly due to the fact that a CB lesson I did a couple weeks ago happened to cover most, if not all of the questions here only. What this means is that I'll need to do far more practice exams to determine my true score.

### Performance Breakdown

| Unit/Topic | Questions Missed | Weakness Identified |
|------------|------------------|---------------------|
| String Manipulation | Q2 (substring operations) | Understanding how substring indices work, especially with concatenation |
| String Methods & Iteration | Q3 (indexOf with loops) | Tracing through complex string operations with while loops |

### What Went Wrong

Looking at the questions I missed, there's a pattern. On the substring question, I didn't carefully trace through what `original.substring(5)` would return ("I") versus `original.substring(1, 2)` (which gives "E"). The answer was B, but I picked C because I miscounted the indices. For the second one with the "lookout" string, I didn't properly trace through the loop iterations to see that it prints substrings *starting with* "o", not immediately after.

### Action Plan for Improvement

**Short-term (Next 2 weeks):**
- Practice 15 minutes daily on substring problems, especially with edge cases
- Create flashcards for String method return values and index behavior
- Trace through at least 3 loop-based string problems per week on paper before coding

**Medium-term (Rest of trimester):**
- Complete one full-length practice MCQ per week
- Review incorrect answers immediately and identify the conceptual gap
- Build a "mistake log" to track recurring weak spots

**Long-term (Exam prep):**
- Target score: 38+/42 on practice exams by March
- Review all College Board FRQ questions involving strings
- Study with classmates who excel at these topics

## (2) Growth Since September

Honestly? I came into this year riding high from CSP and PilotCity. We built Optivize last year—this whole AI-powered small business management platform with stock alerts and everything—and I thought I was going to continue that momentum. Then we started CSA, and Mr. Mortensen had us creating lessons for CSSE students. I'm not gonna lie, I thought it was boring at first. Making markdown tutorials for basic operations? Teaching beginners? It felt like busywork.

But here's what changed: I stopped resisting and started finding the angle that actually mattered to me. Technical communication. I realized these lessons weren't just about teaching loops or conditionals—they were about translating complex ideas into something accessible. That's a skill. After Night at the Museum, when people actually complimented our work and found it genuinely helpful, something clicked. I can apply myself to *anything* if I find the right lens.

So comparing myself to September? I didn't know technical communication could take so many forms. I thought passion only came from building flashy apps. Now I know it can come from teaching, documenting, and making knowledge accessible. That's not something I expected to learn from what I initially dismissed as "boring crap."

## (3) Next Steps for Our Project

Given the time we have left, I want to take the microblog service Mr. Mortensen created and transform it into something bigger. Think Slack, but specifically for Del Norte CS. Full VOIP integration, dedicated channels for different classes and projects, threaded conversations—the whole thing directly integrated into our coding society's workflow.

Yeah, it sounds ambitious. Maybe even far-fetched depending on how the school's Slack situation plays out. But that's exactly the kind of challenge that gets me excited. We already have the foundation with the current messaging system. Now it's about scaling it, adding features that actually serve our community, and creating something that outlasts us.

This trimester taught me that even when projects start small or seem mundane, there's always potential to expand them into something meaningful. That's what I want to carry forward.
