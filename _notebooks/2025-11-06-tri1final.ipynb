### Detailed Error Analysis

**Question 2 - Substring Concatenation:**
- **Target Output:** "HIDE"
- **My Answer:** C - `original.substring(5) + original.substring(1, 2)`
- **Correct Answer:** B - `original.substring(4) + original.substring(0, 2)`
- **Root Cause:** Failed to properly map indices to characters. I assumed substring(5) would give "HI" when it actually returns just "I". **The issue is off-by-one errors and not visualizing the string positions before coding.**

**Question 3 - indexOf Loop Behavior:**
- **Expected Output:** "ookout okout out"
- **My Answer:** D - "kout ut"
- **Correct Answer:** A - "ookout okout out"
- **Root Cause:** Misunderstood the loop logic. The code prints substrings **starting with "o"**, not immediately after. **I need to trace loops step-by-step on paper rather than trying to run them mentally.**
