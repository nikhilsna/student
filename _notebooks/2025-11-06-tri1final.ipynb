### Detailed Error Analysis

![question]({{site.baseurl}}/images/tr1MC/q7.png)


**Question 7 - Substring Concatenation:**
- **Target Output:** "HIDE"
- **My Answer:** C - `original.substring(5) + original.substring(1, 2)`
- **Correct Answer:** B - `original.substring(4) + original.substring(0, 2)`
- **Root Cause:** Failed to properly map indices to characters. I assumed substring(5) would give "HI" when it actually returns just "I". **The issue is off-by-one errors and not visualizing the string positions before coding.**

![question]({{site.baseurl}}/images/tr1MC/q35.png)

**Question 35 - ArrayList Sorting Logic:**
- **Expected Behavior:** Returns `true` if sorted greatest to least, `false` otherwise
- **My Answer:** C - Returns `true` if sorted least to greatest
- **Correct Answer:** D - Returns `true` if sorted greatest to least
- **Root Cause:** Completely reversed the logic. The method checks if `num > prev` and returns `false` when that's true—meaning it's checking that each number is **not greater than** the previous one, which ensures descending order. **I misread the comparison operator and didn't trace through with actual values to verify my understanding.**
